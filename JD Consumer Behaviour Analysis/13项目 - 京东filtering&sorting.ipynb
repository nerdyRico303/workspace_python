# Data Cleaning

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import pyecharts as pye
```

### Step 1 - Define all the datas

```python
df2 = pd.read_excel(r'C:/Users/Asus/Desktop/13_电商行业_京东消费者数据分析/京东消费者分析数据.xlsx')
df2
```

### Step 2 - NaN Deletion

removing all datas that have null values, selected from the table'age_range' and 'city_level'

```python
con1 = df2[df2['age_range'].isnull()].index
con1
```

```python
df2.drop(con1, inplace= True)
df2
```

```python
con2 = df2[df2['city_level'].isnull()].index
con2
```

```python
df2.drop(con2, inplace= True)
df2
```

```python
print(df2['age_range'].isnull().value_counts())
print(df2['city_level'].isnull().value_counts())
```

### Step 3 - Deleting all invalid gender

```python
con3 = df2[df2['gender'] == 'U'].index
con3
```

```python
df2.drop(con3, inplace= True)
df2
```

```python
(df2['gender']== 'U').value_counts()
```

### Step 4 - Deleting the overly small values from the member category

We assume the table 'vip_number' is the refered table, and 0 as the overly small values

```python
con4 = df2[df2['vip_number'] == 0].index
con4
```

```python
df2.drop(con4, inplace= True)
df2
```

```python
(df2['vip_number'] == 0).value_counts()
```

### Step 5 - Replace all negative values from the 'shop_score'

replace with value 0

```python
df2.loc[df2['shop_score'] <= 0,'shop_score'] = 0
df2
```

```python
(df2['shop_score'] <= -1).value_counts()
```
