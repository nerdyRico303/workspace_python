### 3. Product

```python
xaxis1 = df['category'].value_counts().index.tolist()
yaxis1 = df['category'].value_counts().tolist()  
line1 = (
    Line(
        init_opts = opts.InitOpts(theme = ThemeType.DARK)
    )
    .add_xaxis(xaxis1)
    .add_yaxis('Amount',yaxis1)
    .set_global_opts(
        opts.TitleOpts(title='Purchase by Commodity Type'),
        xaxis_opts=opts.AxisOpts(name='Commodity'),
        yaxis_opts=opts.AxisOpts(name='Purchases'),
        toolbox_opts = opts.ToolboxOpts(),
        datazoom_opts = opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
)
line1.render_notebook()
```

```python
xaxis2 = df['brand'].value_counts().index.tolist()
yaxis2 = df['brand'].value_counts().tolist()
line2 = (
    Line(
        init_opts = opts.InitOpts(theme = ThemeType.DARK)
    )
    .add_xaxis(xaxis2)
    .add_yaxis('Amount',yaxis2)
    .set_global_opts(
        opts.TitleOpts(title='Purchases by Brand'),
        xaxis_opts=opts.AxisOpts(name='Brands'),
        yaxis_opts=opts.AxisOpts(name='Purchases'),
        toolbox_opts = opts.ToolboxOpts(),
        datazoom_opts = opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
)
line2.render_notebook()
```

```python
xaxis11 = xaxis1
y_male = []
y_female = []
for i in range(len(xaxis11)):
    male_temp = len(df[(df['category'] == xaxis11[i]) & (df['gender']=='M')])
    female_temp = len(df[(df['category'] == xaxis11[i]) & (df['gender']=='W')])
    y_male.append(male_temp)
    y_female.append(female_temp)
bar5 = (
    Bar(
        init_opts = opts.InitOpts(
            theme = ThemeType.DARK
        )
    )
    .add_xaxis(xaxis11)
    .add_yaxis('男',y_male)
    .add_yaxis('女',y_female)
    .reversal_axis()
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='Commodities Purchased sorted by Gender'
        ),
        xaxis_opts=opts.AxisOpts(name='Commodities'),
        yaxis_opts=opts.AxisOpts(name='Purchases'),
        legend_opts=opts.LegendOpts(
            pos_left='right'
        )
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(
            is_show=False
        )
    )
)
bar5.render_notebook()
```

### 4. Store

```python
xaxis3 = df['shop_category'].value_counts().index.tolist()
yaxis3 = df['shop_category'].value_counts().tolist()
pie1 = (Pie(
    init_opts = opts.InitOpts(
        bg_color = 'black'
    )
)
        .add(
            '',
            [
                list(shop_category)
                for shop_category in zip(xaxis3, yaxis3)
            ]
        )
        .set_global_opts(
            opts.TitleOpts(
                title='Popular Shops based on Purchases',
                pos_left='center',
                title_textstyle_opts=opts.TextStyleOpts(color="#fff")
            ),
            opts.LegendOpts(
                pos_bottom='bottom', 
                textstyle_opts=opts.TextStyleOpts(color="#fff")
            )
        )
        .set_series_opts(
            label_opts=opts.LabelOpts(is_show=False)
        )
)
pie1.render_notebook()
```

```python
tempx = xaxis1
tempy = []
for i in range(len(tempx)):
    avg = df.loc[(df['category'] == tempx[i]), 'shop_score'].mean()
    tempy.append(avg)   
temp = pd.Series((tempy), pd.Index(tempx))
temp = temp.sort_values(ascending=False)
xaxis7 = temp.index.tolist()
yaxis7 = temp.tolist()
bar1 = (
    Bar(
        init_opts = opts.InitOpts(theme = ThemeType.DARK)
    )
    .add_xaxis(xaxis7)
    .add_yaxis('Average',yaxis7)
    .set_global_opts(
        opts.TitleOpts(title='Popular Shops by Commodity'),
        xaxis_opts=opts.AxisOpts(name='Category'),
        yaxis_opts=opts.AxisOpts(name='Shop Score'),
        toolbox_opts = opts.ToolboxOpts(),
        datazoom_opts = opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
)
bar1.render_notebook()
```

```python
tempx = xaxis2
tempy = []
for i in range(len(tempx)):
    avg2 = df.loc[(df['brand'] == tempx[i]), 'shop_score'].mean()
    tempy.append(avg2)
temp = pd.Series((tempy), pd.Index(tempx))
temp = temp.sort_values(ascending=False)
xaxis8 = temp.index.tolist()
yaxis8 = temp.tolist()
bar2 = (
    Bar(
        init_opts = opts.InitOpts(theme = ThemeType.DARK)
    )
    .add_xaxis(xaxis8)
    .add_yaxis('Average',yaxis8)
    .set_global_opts(
        opts.TitleOpts(title='Popular Shops by Brand'),
        xaxis_opts=opts.AxisOpts(name='Brands'),
        yaxis_opts=opts.AxisOpts(name='Shop Score'),
        toolbox_opts = opts.ToolboxOpts(),
        datazoom_opts = opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False)
    )
)
bar2.render_notebook()
```

```python
xaxis12 = []
for i in range (8):
    xaxis12.append(i)
df['shop_register_date'] = pd.to_datetime(df['shop_register_date'])   
df['shop_register_year'] = df['shop_register_date'].dt.year
yaxis12 = df['shop_register_year'].value_counts()
yaxis12 = yaxis12.sort_index(ascending = False).tolist()
bar = (
    Bar(
        init_opts = opts.InitOpts(theme = ThemeType.DARK)
    )
    .add_xaxis(xaxis12)
    .add_yaxis('Sales',yaxis12)
    .set_global_opts(
        opts.TitleOpts(title='Sales Ammount by Stores age'),
        xaxis_opts=opts.AxisOpts(name='Stores Age'),
        yaxis_opts=opts.AxisOpts(name='Sales'),
        toolbox_opts = opts.ToolboxOpts(),
        legend_opts=opts.LegendOpts(
            is_show= False
        )
    )
)
bar.render_notebook()
```
